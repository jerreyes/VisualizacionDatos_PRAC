# Visualización de datos - Practica 2: Creación de visualización

```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
# Se leerán los datos a utilizar para la visualización

data = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
data.head(10)
```

```python
# Se obtienen las diferentes features que compenen el set de datos
data.columns
```

```python
# Se obtiene la longitud total del set de datos y el número de features
print(data.shape)
```

```python
# Obtenemos los números de los valores nulos para cada columna en el dataset
missing_values_count = data.isnull().sum() 
missing_values_count
```

```python
data_drop = data.drop(columns=['region_1', 'region_2', 'designation', 'province'])
print(data_drop.shape)
print("\n\n\n")
data_drop.head(20)
```

```python
# Se vuelve a comprobar los valores nulos que se obtienen
missing_values_datadrop = data_drop.isnull().sum() 
missing_values_datadrop
```

```python
# Se observa que aun tenemos valores nulos en diversas variables, en este siguiente paso se procederá a imputar la feature price
# Se obtiene la media sin decimales, redondeada al valor entero
mean_value = round(data_drop['price'].mean(), 0)
print(mean_value)
```

```python
data_drop['price'].fillna(value=mean_value, inplace=True)
print('Dataframe actualizado:\n')
data_drop.head(10)
```

```python
# Volvemos a comprobar valores nulos
missing_values_datadrop = data_drop.isnull().sum() 
missing_values_datadrop
```

A continuación se procederá a eliminar aquellas filas que tengan valores nulos, ya se advierte que como minimo perderemos 31213 filas, ya que la variable 'taster_twitter_handle', esto se realiza de esta manera porque se pretende utilizar dicha variable para futuras representaciones en tableau si es posible. Tras realizar esta accion tendremos el dataset sin ningún tipo de valor nulo y podremos crear alguna nueva variable para utilizar más adelante en la representación.

```python
# Eliminamos todas aquellas filas que tengan un valor nulo
data_clean = data_drop.dropna()
data_clean.head(10)
```

```python
# Se comprueba que no existen valores nulos en el dataset
data_clean.isnull().sum()
```

```python
# Se observa la distribución de la variable númerica price

data.price.describe()
```

```python
data_clean.loc[data_clean.price < 10,'price_cat']='Very cheap'
data_clean.loc[(data_clean.price > 10) & (data_clean.price < 25) ,'price_cat']='Cheap'
data_clean.loc[(data_clean.price > 25) & (data_clean.price < 40) ,'price_cat']='Normal'
data_clean.loc[(data_clean.price > 40) & (data_clean.price < 300) ,'price_cat']='Expensive'
data_clean.loc[data_clean.price > 300 ,'price_cat']='Luxury'
```
